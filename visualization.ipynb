```python
import pandas as pd
from ggplot import *
import numpy as np
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
for name,frame in df.groupby("Embarked"):
    print ggplot(aes(x="Survived"),data=frame) + geom_histogram() + ggtitle("Embarked class: "+str(name))
```

```python
for name,frame in df.groupby("Sex"):
    print ggplot(aes(x="Survived"),data=frame) + geom_histogram() + ggtitle("Gender: "+str(name))
```
